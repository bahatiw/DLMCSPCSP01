# Import libraries we need for this project

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn import preprocessing
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
from sklearn.metrics import confusion_matrix, accuracy_score
import sklearn.metrics as metrics
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from time import time
```

### Specify the file contain our data.

```python
filename ="Weather_Data.csv"
```

### Read This into a Panda DataFrame

```python
df = pd.read_csv("Weather_Data.csv")
```

```python
print('Loaded {} records from {}.\n'.format(len(df), filename))
```

### Check that we have loaded the correct data

```python
df.head()
```

### For Regression we need to convert some categorical data to binary in order for the model to work. In our case RainToday,WindGustDir,WindDir9am,WindDir3pm. We perform one hot encoding 

```python
df_processed = pd.get_dummies(data=df, columns=['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
```

## Reconfirm our data is in shape

```python
df_processed.head()
```

### We also change the RainTommorow Column to binary from "Yes" and "No" for regression

```python
df_processed.replace(['No', 'Yes'], [0,1], inplace=True)
```

# Prepare our data for Training and Testing

```python
df_processed.drop('Date',axis=1,inplace=True)
```

```python
df_processed = df_processed.astype(float)
```

```python
x_values = df_processed.drop(columns='RainTomorrow', axis=1)
```

```python
y_values = df_processed['RainTomorrow']
```

# Decision Tree 

### We split the data for training and testing set to make it 20% for testing and 80% for Training

```python
x_train, x_test, y_train, y_test =  train_test_split(x_values,y_values , random_state=10,test_size=0.2, shuffle=True)
```

```python
tree = DecisionTreeClassifier(random_state = 1912)
start = time()
tree.fit(x_train, np.ravel(y_train))
end = time()
train_time = (end - start) * 1000

prediction = tree.predict(x_test)

# Score using the validation data.
score = tree.score(x_test, y_test)

print('Decision tree model took {:.2f} milliseconds to fit.'.format(train_time))
print('Accuracy: {:.0f}%'.format(score * 100))
```

## Visualize the decision tree structure

## Compute accuracy, precision, recall, and F1 score

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

def model_scores(y, prediction):
    acc = accuracy_score(y, prediction)
    print('Accuracy: {:.0f}%'.format(np.round(acc * 100)))
    
    precision = precision_score(y, prediction)
    print('Precision: {:.0f}%'.format(np.round(precision * 100)))
    
    recall = recall_score(y, prediction)
    print('Recall: {:.0f}%'.format(np.round(recall * 100)))
    
    f1 = f1_score(y, prediction)
    print('F1: {:.0f}%'.format(np.round(f1 * 100)))
    
print('A function to compute the model scores has been defined.')
```

## Generate a ROC curve and compute the AUC

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

def roc(y, prediction_proba):
    fpr, tpr, thresholds = roc_curve(y, prediction_proba)
    
    plt.plot(fpr, tpr);
    plt.xlim([0.0, 1.0]);
    plt.ylim([0.0, 1.0]);
    plt.title('ROC Curve');
    plt.xlabel('False Positive Rate');
    plt.ylabel('True Positive Rate');
    plt.grid(True);
    
    auc = roc_auc_score(y, prediction_proba)
    print('Area Under Curve: {:.2f}'.format(auc))
    
print('A function to generate the ROC curve and compute AUC has been defined.')
```

## Generate a precisionâ€“recall curve and compute the average precision

```python
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import average_precision_score

def prc(y, prediction_proba):
    precision, recall, thresholds = precision_recall_curve(y, prediction_proba)
    
    plt.plot(recall, precision);
    plt.xlim([0.0, 1.0]);
    plt.ylim([0.0, 1.0]);
    plt.title('Precisionâ€“Recall Curve');
    plt.xlabel('Recall');
    plt.ylabel('Precision');
    plt.grid(True);
    
    ap = average_precision_score(y, prediction_proba)
    print('Average Precision: {:.2f}'.format(ap))
    
print('A function to generate the PRC and compute average precision has been defined.')
```

## Evaluate the initial decision tree model

```python
initial_predict = tree.predict(x_test)

model_scores(y_test, initial_predict)
```

```python
initial_predict_proba = tree.predict_proba(x_test)

roc(y_test, initial_predict_proba[:, 1])
```

```python
prc(y_test, initial_predict_proba[:, 1])
```

## Performing pre-pruning on the decision tree

```python
from sklearn.tree import DecisionTreeClassifier

pruned_tree = DecisionTreeClassifier(max_depth = 4, random_state = 1912)
start = time()
pruned_tree.fit(x_train, np.ravel(y_train))
end = time()
train_time = (end - start) * 1000

prediction = pruned_tree.predict(x_test)

# Score using the validation data.
score = pruned_tree.score(x_test,y_test)

print('Decision tree model took {:.2f} milliseconds to fit.'.format(train_time))
print('Accuracy: {:.0f}%'.format(score * 100))
```

## Evaluate the pruned decision tree modelÂ¶

```python
pruned_predict_proba = pruned_tree.predict_proba(x_test)

roc(y_test, pruned_predict_proba[:, 1])
```

```python
prc(y_test, pruned_predict_proba[:, 1])
```

## Fit a decision tree model using randomized search with cross-validation

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint as sp_randint

dist = {'criterion': ['gini', 'entropy'],
        'splitter': ['best', 'random'],
        'max_depth': sp_randint(2, 10),
        'min_samples_split': sp_randint(5, 100),
        'min_samples_leaf': sp_randint(5, 100)}

search = RandomizedSearchCV(tree, param_distributions = dist, n_iter = 500, 
                            scoring = 'f1', cv = 5, random_state = 1912)
search.fit(x_train, np.ravel(y_train));
optimized_tree = search.best_estimator_

print(search.best_params_)
```

```python
optimized_predict = optimized_tree.predict(x_test)

model_scores(y_test, optimized_predict)
```

```python
optimized_predict_proba = optimized_tree.predict_proba(x_test)

roc(y_test, optimized_predict_proba[:, 1])
```

```python
prc(y_test, optimized_predict_proba[:, 1])
```
